```python
#lib
from bs4 import BeautifulSoup
import requests
```

```python
changiai_airport_url = "https://www.changiairport.com/in/en.html"
changiai_airport_response = requests.get(changiai_airport_url)
changiai_airport_soup = BeautifulSoup(changiai_airport_response.text, 'lxml')
```
